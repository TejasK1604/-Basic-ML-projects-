```python
df.drop(columns=["Unnamed: 0","Email","Address","Avatar"], inplace=True)
```

```python
x= df.drop('Yearly Amount Spent', axis=1)
y= df['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=40)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
LinearRegression
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
coeff_df = pd.DataFrame(lr.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
pred_train_lr= lr.predict(x_train)
print('Training data (RMSE):', np.sqrt(mean_squared_error(y_train, pred_train_lr)))
print('Training data (r2 score):', r2_score(y_train, pred_train_lr))
```

```python
pred_test_lr= lr.predict(x_test)
print('Training data (RMSE):', np.sqrt(mean_squared_error(y_test, pred_test_lr)))
print('Training data (r2 score):',r2_score(y_test, pred_test_lr))
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
rr= Ridge(alpha=1)
rr.fit(x_train, y_train)
```

```python
rr_coeff= pd.DataFrame(rr.coef_,x.columns, columns=['Coffecient'])
rr_coeff
```

```python
pred_train_rr= rr.predict(x_train)
print('Training data (RMSE):', np.sqrt(mean_squared_error(y_train, pred_train_rr)))
print('Training data (r2 score):', r2_score(y_train, pred_train_rr))
```

```python
pred_test_rr= rr.predict(x_test)
print('Training data (RMSE):', np.sqrt(mean_squared_error(y_test, pred_test_rr)))
print('Training data (r2 score):',r2_score(y_test, pred_test_rr))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

RidgeRegression= Ridge()
hyperParameters= {'alpha': [1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridgeRegressor= GridSearchCV(RidgeRegression, hyperParameters, cv=5)
ridgeRegressor.fit(x_train, y_train)
```

```python
print("Best Value for lambda: ", ridgeRegressor.best_params_)
print("Best Score for cost function: ", ridgeRegressor.best_score_)
```

```python
rr1= Ridge(alpha=0.01)
rr1.fit(x_train, y_train)
```

```python
pred_train_rr1= rr1.predict(x_train)
print('Training data (RMSE):', np.sqrt(mean_squared_error(y_train, pred_train_rr1)))
print('Training data (r2 score):',r2_score(y_train, pred_train_rr1))
```

```python
pred_test_rr1= rr1.predict(x_test)
print('Training data (RMSE):', np.sqrt(mean_squared_error(y_test, pred_test_rr1)))
print('Training data (r2 score):',r2_score(y_test, pred_test_rr1))
```
